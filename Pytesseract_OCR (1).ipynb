### Explanation of Changes:

1. **Custom Configuration for Digit Recognition**:
   - `custom_config = r'--oem 3 --psm 6 outputbase digits'`: This line sets the Tesseract configuration to recognize only digits.
     - `--oem 3`: Specifies the OCR Engine Mode. `3` means both the legacy Tesseract OCR engine and the LSTM OCR engine are used.
     - `--psm 6`: Specifies the Page Segmentation Mode. `6` means Tesseract assumes a single uniform block of text.
     - `outputbase digits`: This option tells Tesseract to recognize only digits.

2. **Passing the Custom Configuration to Tesseract**:
   - `text = pytesseract.image_to_string(gray, config=custom_config)`: This line performs OCR using Tesseract with the custom configuration set for digit recognition.

###Whitelisting and Blacklisting

```python
# Import the necessary packages
import pytesseract
import cv2
import re
from google.colab import files

# Function to upload an image file
def upload_image():
    uploaded = files.upload()
    if len(uploaded) > 0:
        image_path = next(iter(uploaded.keys()))
        return image_path
    else:
        raise Exception("No image uploaded")

# Upload the image file
image_path = upload_image()

# Load the image using OpenCV
image = cv2.imread(image_path)

# Convert the image to grayscale (optional, but recommended)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to preprocess the image (optional, but can improve OCR accuracy)
gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

# Save the preprocessed image (optional)
preprocessed_image_path = "preprocessed_image.png"
cv2.imwrite(preprocessed_image_path, gray)

# Initialize options string
options = ""

# Define whitelist and blacklist characters
whitelist = "0123456789"  # Example: Only allow digits
blacklist = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"  # Example: Disallow all alphabetic characters

# Check to see if a set of whitelist characters has been provided,
# and if so, update our options string
if len(whitelist) > 0:
    options += "-c tessedit_char_whitelist={}".format(whitelist)

# Check to see if a set of blacklist characters has been provided,
# and if so, update our options string
if len(blacklist) > 0:
    options += "-c tessedit_char_blacklist={}".format(blacklist)

# Perform OCR on the preprocessed image with the configured options
text = pytesseract.image_to_string(gray, config=options)

# Print the extracted text to understand the effect of whitelist and blacklist
print("Extracted Text with Whitelist and Blacklist:")
print(text)

# Define the pattern for a CNIC number (XXXXX-XXXXXXX-X)
cnic_pattern = re.compile(r'\b\d{5}-\d{7}-\d{1}\b')

# Search for the CNIC pattern in the extracted text
cnic_numbers = cnic_pattern.findall(text)

# Print the extracted CNIC numbers
print("Extracted CNIC Numbers:")
for cnic in cnic_numbers:
    print(cnic)
```

###Orientation and Script Detection

```python
# Import the necessary packages
import pytesseract
import cv2
from google.colab import files

# Function to upload an image file
def upload_image():
    uploaded = files.upload()
    if len(uploaded) > 0:
        image_path = next(iter(uploaded.keys()))
        return image_path
    else:
        raise Exception("No image uploaded")

# Upload the image file
image_path = upload_image()

# Load the image using OpenCV
image = cv2.imread(image_path)

# Perform orientation and script detection
# The config parameter '-l osd --psm 0' is used for orientation and script detection
osd = pytesseract.image_to_osd(image)
print(osd)
# Parse the orientation and script detection results
def parse_osd(osd_data):
    result = {}
    for line in osd_data.split('\n'):
        if 'Orientation in degrees' in line:
            result['Orientation'] = int(line.split(':')[-1].strip())
        elif 'Rotate' in line:
            result['Rotate'] = int(line.split(':')[-1].strip())
        elif 'Orientation confidence' in line:
            result['Orientation confidence'] = float(line.split(':')[-1].strip())
        elif 'Script' in line:
            result['Script'] = line.split(':')[-1].strip()
        elif 'Script confidence' in line:
            result['Script confidence'] = float(line.split(':')[-1].strip())
    return result

# Get the parsed results
parsed_osd = parse_osd(osd)

# Print the results
print("Orientation and Script Detection Results:")
for key, value in parsed_osd.items():
    print(f"{key}: {value}")
```

###PSM Modes

```python
from google.colab import files
import cv2
import pytesseract

# Upload the image file
uploaded = files.upload()

# Assuming only one file is uploaded
image_path = next(iter(uploaded.keys()))

# Load the image using OpenCV
image = cv2.imread(image_path)

# Convert the image to grayscale (optional, but recommended)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply thresholding to preprocess the image (optional, but can improve OCR accuracy)
gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

# Save the preprocessed image (optional)
preprocessed_image_path = "preprocessed_image.png"
cv2.imwrite(preprocessed_image_path, gray)

psm_modes = {
    0: "Orientation and script detection (OSD) only.",
    1: "Automatic page segmentation with OSD.",
    2: "Automatic page segmentation, but no OSD or OCR.",
    3: "Fully automatic page segmentation, but no OSD.",
    4: "Assume a single column of text of variable sizes.",
    5: "Assume a single uniform block of vertically aligned text.",
    6: "Assume a single uniform block of text.",
    7: "Treat the image as a single text line.",
    8: "Treat the image as a single word.",
    9: "Treat the image as a single word in a circle.",
    10: "Treat the image as a single character.",
    11: "Sparse text. Find as much text as possible in no particular order.",
    12: "Sparse text with OSD.",
    13: "Raw line. Treat the image as a single text line."
}

for psm in psm_modes:
    try:
        custom_config = r'--psm {}'.format(psm)
        text = pytesseract.image_to_string(gray, config=custom_config)
        print(f"PSM Mode {psm}: {psm_modes[psm]}")
        print("Extracted Text:")
        print(text)
        print("\n" + "="*40 + "\n")
    except pytesseract.TesseractError as e:
        print(f"Error with PSM Mode {psm}: {e}")
    except Exception as e:
        print(f"An unexpected error occurred with PSM Mode {psm}: {e}")
```
